## This script is for loading and processing data, which come out of the Ipfixprobe when fed a simple .pcap file

**Import everything needed for data processing**

```python
import pandas as pd
import numpy as np
```

**Load The CSV**    
The CSV is loaded from Google Drive

```python
loaded_csv = pd.read_csv('/content/drive/MyDrive/Omega_Data/big_data.csv')
```

**Create The DataFrame**

```python
df = pd.DataFrame(loaded_csv)
df
```

**Extract and rename the columns we want to work with**

```python
df.columns
```

```python
df = df[['ipaddr DST_IP', 'ipaddr SRC_IP', 'int8* PPI_PKT_DIRECTIONS', 'uint16* PPI_PKT_LENGTHS']]
df = df.rename(columns={'ipaddr DST_IP': "DST_IP", 'ipaddr SRC_IP': "SRC_IP", 'int8* PPI_PKT_DIRECTIONS': "PKT_DIRECTIONS", "uint16* PPI_PKT_LENGTHS": "PKT_LENGTHS"})
```

```python
df
```

**Transform pkt_directions into columns and join it with the dataframe**

```python
df = df.join(df['PKT_DIRECTIONS'].str.replace("[", "").str.replace("]", "").str.split("|", expand=True)).drop('PKT_DIRECTIONS', axis=1)
df
```

**Drop columns where the first byte direction is an empty string. That means that all packets in the communication had size of 0, which we dont need**

```python
df = df.drop(df[0][df[0] == ''].index)
df
```

**Similarly to packet directions, we will split the packet lengths into several columns. This time we can cast it to integers, because we no longer have those empty string rows**

```python
pkt_lengths = df['PKT_LENGTHS'].str.replace("[", "").str.replace("]", "").str.split("|", expand=True).replace(to_replace=[None], value=0).astype(int)
pkt_lengths
```

**Last step is to multiply packet directions with packet lengths and join it with the src_ip and dst_ip**

```python
final_dataframe = df[['SRC_IP', 'DST_IP']].join(df.iloc[:, 3:].replace(to_replace=[None], value=0).astype(int)*pkt_lengths)
final_dataframe
```

**We can export the intermidiate dataframe**

```python
final_dataframe.to_csv("final_dataframe.csv")
```

# Exctract rows with public IPs, which we will use for annotation

**Import everything needed for reverse dns lookup**

```python
from ipaddress import ip_address
```

**Extract the source ip and destination ip from our dataframe**

```python
ip_dataframe = final_dataframe[['SRC_IP', 'DST_IP']]
ip_dataframe
```

**Extract public ip addresses from source ips. Private addresses will be changed to NaN**

```python
filter = ip_dataframe['SRC_IP'].apply(lambda x: ip_address(x).is_global)
src_ip_global = ip_dataframe['SRC_IP'].where(filter)
src_ip_global
```

**Extract public ip addresses from destination ips. Private addresses will be changed to NaN**

```python
filter2 = ip_dataframe['DST_IP'].apply(lambda x: ip_address(x).is_global)
dst_ip_global = ip_dataframe['DST_IP'].where(filter2)
dst_ip_global
```

**Join source public ips with destination public ips into one column. We know that there never can be a public address both in source and destination ip, so we use fillna() method to replace the NaN value with the ip address if possible. There can be rows where there were two private ip addresses. Those will stay NaN**

```python
src_dst_global_ips = src_ip_global.fillna(dst_ip_global)
src_dst_global_ips = src_dst_global_ips.rename("IP")
src_dst_global_ips
```

**We will join the newly created column with our final dataframe**

```python
final_dataframe = final_dataframe.join(src_dst_global_ips)
final_dataframe
```

**We can now drop the SRC_IP and DST_IP columns**

```python
final_dataframe = final_dataframe.drop(['SRC_IP', 'DST_IP'], axis=1)
final_dataframe
```

**We will now also drop the rows, where no public ip address was found**

```python
final_dataframe = final_dataframe.drop(final_dataframe[final_dataframe["IP"].isna()].index)
final_dataframe
```

**We can export the intermidiate dataframe**

```python
final_dataframe.to_csv("final_dataframe_ips.csv")
```

# Perform reverse dns lookup of pubic ips and create final annotated dataset

```python
import socket
import re
```

**We will create a simple function which will try to resolve the ip address. If no domain is found, None is returned**

```python
def resolve(ip):
  try:
    return socket.gethostbyaddr(ip)[0]
  except:
    return None
domains = final_dataframe['IP'].apply(lambda x: resolve(x))
domains
```

**We will replace our IP column with the resolved domains**

```python
final_dataframe['IP'] = domains
final_dataframe
```

**We will drop the rows where the public ip was not resolved and rename the column to Service**

```python
final_dataframe = final_dataframe.drop(final_dataframe[final_dataframe['IP'].isnull()].index)
final_dataframe = final_dataframe.rename(columns={"IP": "Service"})
final_dataframe
```

**We can now export the services to look for interesting domain names**

```python
final_dataframe['Service'].to_csv('services.csv')
```

**We will now specify our classes. The dictionary works like this: The key is the keyword we are looking for in the domain and the value is the class we will replace it with**

```python
classes = {'github': 'github', 'facebook': 'facebook', 'seznam': 'seznam'}
```

**We will create a function which will automatically assign classes to domains and drop the domains which were not classified**

```python
def classify(annotations):
  for k, v in classes.items():
    final_dataframe['Service'] = final_dataframe['Service'].apply(lambda x: v if k in x else x)
  return final_dataframe[final_dataframe['Service'].isin(classes.values())]
final_dataframe_with_classes = classify(classes)
final_dataframe_with_classes
```

**Export the final dataset with classes**

```python
final_dataframe_with_classes.to_csv('final_dataframe_with_classes.csv')
```
