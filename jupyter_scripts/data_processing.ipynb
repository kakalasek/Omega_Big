# This script processes the data into a dataset usable by the training algorithm

## Import pandas and numpy

```python
import pandas as pd
import numpy as np
```

## Read our data from csv to pandas datafram

```python
df = pd.read_csv('../data/tadas_and_minivan_aggregated.csv')
df
```

## Extract only rows with destination port of 443

```python
df = df[df['uint16 DST_PORT'] == 443]
df
```

## Remove duplicate rows

```python
df = df.drop_duplicates()
df
```

## See our columns

```python
df.columns
```

## Choose the columns we need       
**Which is *TLS_SNI*, which contains our domain (service) and will be used for annotattion, *PPI_PKT_DIECTIONS*, which contains 1 or -1 for each packet in the flow depending on the direction of the packet, and *PPI_PKT_LENGTHS*, which contains the lengths of first 30 packets**           
**We will also rename the columns so the work with them will be easier**

```python
df = df[['string TLS_SNI', 'int8* PPI_PKT_DIRECTIONS', 'uint16* PPI_PKT_LENGTHS']]
df = df.rename(columns={'string TLS_SNI': 'TLS_SNI', 'int8* PPI_PKT_DIRECTIONS': "PKT_DIRECTIONS", "uint16* PPI_PKT_LENGTHS": "PKT_LENGTHS"})
df
```

## Split packet directions into 30 independent columns, each containing either None (No packet there) or 1 or -1

```python
df = df.join(df['PKT_DIRECTIONS'].str.replace("[", "").str.replace("]", "").str.split("|", expand=True)).drop('PKT_DIRECTIONS', axis=1)
df
```

## We now need to drop the packets, which have an empty string in the first packet direction, because that means, that there is no packet in the flow

```python
df = df.drop(df[0][df[0] == ''].index)
df
```

## Then we will create a separate dataframe containing the length of the first 30 packets

```python
pkt_lengths = df['PKT_LENGTHS'].str.replace("[", "").str.replace("]", "").str.split("|", expand=True).replace(to_replace=[None], value=0).astype(int)
pkt_lengths
```

## Now we will replace the None values in our packet directions with zeros and multiply the with our packet lengths, thus finally getting a dataframe with 30 packet lengths and directions in one.

```python
df = df[['TLS_SNI']].join(df.iloc[:, 2:].replace(to_replace=[None], value=0).astype(int)*pkt_lengths)
df
```

## Now we want to drop the packets, which have NaN value in the TLS_SNI field, because they were probably not flow to outside world and we could not properly annonate them

```python
df = df.drop(df[df['TLS_SNI'].isna()].index)
df
```

## We will create a new column called 'class' and assign a default value of 'other' to every row            
**This 'class' column will contain our annotations**

```python
df['class'] = 'other'
df
```

## We can now export our TLS_SNI to csv, so we can see, which domains are present and wheter we would like to annotate them

```python
df['TLS_SNI'].to_csv('domains.csv')
```

## We will now define the services we want to classify inside a python dictionary and classify them based on the presence of the dict key inside a TLS_SNI field

**Load our classes**

```python
import json

classes = {}
with open('classes.json') as file:
    classes = json.load(file)

classes
```

**Assign classes to flows we can**

```python
for k, v in classes.items():
    df.loc[df['TLS_SNI'].str.contains(k), 'class'] = v

df
```

## We will now load our json for mappping numerical value to each service

```python
classes_mapping = {}
with open('classes_mapping.json') as file:
    classes_mapping = json.load(file)

classes_mapping
```

## Finally we can replace the classnames in our 'class' column with the numerical values

```python
for k, v in classes_mapping.items():
  df.loc[df['class'] == v, 'class'] = k

df
```

## Now we can drop the TLS_SNI column

```python
df.drop(columns=['TLS_SNI'], inplace=True)
df
```

## Export our final dataset

```python
df.to_csv('final_dataset.csv')
```
