```python
df = df[['string TLS_SNI', 'int8* PPI_PKT_DIRECTIONS', 'uint16* PPI_PKT_LENGTHS']]
df = df.rename(columns={'string TLS_SNI': 'TLS_SNI', 'int8* PPI_PKT_DIRECTIONS': "PKT_DIRECTIONS", "uint16* PPI_PKT_LENGTHS": "PKT_LENGTHS"})
df
```

```python
df = df.join(df['PKT_DIRECTIONS'].str.replace("[", "").str.replace("]", "").str.split("|", expand=True)).drop('PKT_DIRECTIONS', axis=1)
df
```

```python
df = df.drop(df[0][df[0] == ''].index)
df
```

```python
pkt_lengths = df['PKT_LENGTHS'].str.replace("[", "").str.replace("]", "").str.split("|", expand=True).replace(to_replace=[None], value=0).astype(int)
pkt_lengths
```

```python
df = df[['TLS_SNI']].join(df.iloc[:, 2:].replace(to_replace=[None], value=0).astype(int)*pkt_lengths)
df
```

```python
df = df.drop(df[df['TLS_SNI'].isna()].index)
df
```

```python
df['class'] = 'other'
df
```

```python
classes = {'github': 'github', 'facebook': 'facebook', 'seznam': 'seznam', 'spotify': 'spotify', 'fb.com': 'facebook', 'edu.cz': 'edu.cz', 'outlook': 'outlook', 'mail.google': 'gmail'}

for k, v in classes.items():
    df.loc[df['TLS_SNI'].str.contains(k), 'class'] = v

df[df['class'] == 'gmail']
```

```python
df['TLS_SNI'].to_csv('domains.csv')
```

```python
def generate_classes(data_frame, column_name):
  output_dict = {}
  services = []
  for service in [x for x in data_frame[column_name]]:
    if service not in services:
      services.append(service)

  current_annotation = 1

  while current_annotation <= len(services):
    output_dict[current_annotation] = services[current_annotation-1]
    current_annotation += 1

  return output_dict

classes_mapping = generate_classes(df, 'class')
classes_mapping
```

```python
for k, v in classes_mapping.items():
  df.loc[df['class'] == v, 'class'] = k

df
```
